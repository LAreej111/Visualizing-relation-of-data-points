#  Q.2: Check statistical facts by checking all columns. Then calculate the maximum value of the following attributes ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']

```python
import pandas as pd

# Assuming you have a DataFrame named 'nutrition_data' containing the nutritional information
# Load your dataset
# Example:
nutrition_data = pd.read_csv('menu.csv')

# Check statistical facts for all columns
statistical_facts = nutrition_data.describe()

# List of attributes
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
              'Iron (% Daily Value)']

# Calculate the maximum value for each attribute
max_values = {}
for attribute in attributes:
    max_value = nutrition_data[attribute].max()
    max_values[attribute] = max_value

print("Maximum values for each attribute:")
for attribute, value in max_values.items():
    print(f"{attribute}: {value}")
```

# Q.3: Check to see if infact there is any correlation between Calories and other independent variables by plotting a correlation matrix next

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'nutrition_data' containing the nutritional information
# Load your dataset
# Example:
nutrition_data = pd.read_csv('menu.csv')

# Calculate the correlation matrix
correlation_matrix = nutrition_data.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Nutritional Information')
plt.show()
```

We calculate the correlation matrix using the corr() method provided by pandas DataFrame. This method computes pairwise correlation of columns, excluding NA/null values.
We then use seaborn's heatmap() function to create a heatmap of the correlation matrix. The argument annot=True adds the correlation coefficients to the heatmap, cmap='coolwarm' sets the color palette, and fmt=".2f" formats the annotation values to two decimal places.

# Q.5: Figure out which exact item contains a high quantity for ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)'.

```python
import pandas as pd

# Assuming you have a DataFrame named 'nutrition_data' containing the nutritional information
# Load your dataset
# Example:
nutrition_data = pd.read_csv('menu.csv')

# List of attributes
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
              'Iron (% Daily Value)']

# Dictionary to store the item with maximum value for each attribute
max_items = {}

# Iterate over each attribute
for attribute in attributes:
    # Find the row with maximum value for the current attribute
    max_row = nutrition_data.loc[nutrition_data[attribute].idxmax()]
    # Extract the item name
    item_name = max_row['Item']
    # Store the item name and its corresponding attribute value in the dictionary
    max_items[attribute] = {'Item': item_name, 'Value': max_row[attribute]}

# Display summary
print("Summary of Items with Maximum Values for Each Attribute:")
for attribute, info in max_items.items():
    print(f"{attribute}: {info['Item']} - {info['Value']}")
```

#  Q.6: Draw Stripplot for each category against the following attributes ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein','Vitamin A (% Daily Value)','Vitamin C (% Daily Value)','Calcium (% Daily Value)', 'Iron (% Daily Value)'. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'nutrition_data' containing the nutritional information
# Load your dataset
# Example:
nutrition_data = pd.read_csv('menu.csv')

# List of attributes
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
              'Iron (% Daily Value)']

# Set up the figure size
plt.figure(figsize=(15, 10))

# Iterate over each attribute
for i, attribute in enumerate(attributes, 1):
    # Create subplot for each attribute
    plt.subplot(3, 4, i)
    # Draw stripplot
    sns.stripplot(x='Category', y=attribute, data=nutrition_data, jitter=True)
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=90)
    # Set title for each subplot
    plt.title(f'{attribute} Distribution Across Categories')

# Adjust layout
plt.tight_layout()
plt.show()
```

We iterate over each attribute and create a subplot for each attribute using plt.subplot().
We draw a stripplot for each attribute against the 'Category' column from the dataset using sns.stripplot().
The jitter=True parameter adds a small amount of random noise to separate the points along the categorical axis, which helps prevent overplotting.
We rotate the x-axis labels to ensure better readability.
Finally, we use plt.tight_layout() to adjust the layout of the subplots and plt.show() to display the plot.

# Q.7: Draw a horizontal bar graph for items in each category against the calories. 


```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'nutrition_data' containing the nutritional information
# Load your dataset
# Example:
nutrition_data = pd.read_csv('menu.csv')

# Group the data by category and find the mean calories for each category
category_calories_mean = nutrition_data.groupby('Category')['Calories'].mean().sort_values()

# Create the horizontal bar graph
plt.figure(figsize=(10, 8))
category_calories_mean.plot(kind='barh', color='purple')

# Set labels and title
plt.xlabel('Mean Calories')
plt.ylabel('Category')
plt.title('Mean Calories for Items in Each Category')

# Show the plot
plt.show()
```
